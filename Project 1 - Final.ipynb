# Concatenate UC Data CSV's

```python
#Merging the data of all 5 years of public and private school data using concat

All_complete_data = pd.concat([uc_data_complete_2020, nine_data_complete,
                               eight_data_complete, seven_data_complete, six_data_complete])
#Re-index data frame
All_complete_data = All_complete_data.reset_index(drop=True)
All_complete_data
```

```python
#Groupby series to iterate through Count row to see if App is within list, then apply filter if true; iloc to keep indices
All_complete_data = All_complete_data.iloc[All_complete_data.groupby(["Calculation1", "Year"])\
                                .filter(lambda x: "App" in x["Count"].to_list()).index]
```

```python
#Creating the percent of Applicant Pool column, compare Admission & Enrollment rates across schools
All_complete_data["Application Total"] = All_complete_data.groupby(["Year", "Calculation1"])["All"].transform("max")
All_complete_data["Percent of Applicant Pool"] = All_complete_data["All"]/All_complete_data["Application Total"]*100
```

# Join CA High Schools w/ County Median Income (for top/bottom analysis)

```python
county_income = "County income.xlsx"
plus_income = pd.read_excel(county_income)
plus_income
```

```python
#Check county list within County Median Income data frame
plus_income["County"].unique()
```

```python
#Check county list within UC complete data frame
sorted(All_complete_data["County/State/ Territory"].unique())
```

```python
#Merge complete UC data and County Income
all_complete_new = All_complete_data.merge(plus_income, how = "left", left_on = "County/State/ Territory", right_on = "County")

#Calculate average median income of CA counties
avg_med_income = all_complete_new["Value"].mean()

all_complete_new["Value Minus Average Median Income"] = all_complete_new["Value"] - avg_med_income
all_complete_new
```

# Look at % of Public/Private with Admission over 80% (compared to applied)

```python
#Count of 80%+ acceptance of public school applicants (into the UC System)
public_adm_count = all_complete_new[(all_complete_new["Count"] == "Adm")
                  & (all_complete_new["Percent of Applicant Pool"] >= 80)
                  & (all_complete_new["Institution_type"] == "Public")]["Calculation1"].count()

total_public_count = all_complete_new[(all_complete_new["Count"] == "App") &
                  (all_complete_new["Institution_type"] == "Public")]["Calculation1"].count()

public_adm_count/total_public_count*100
print("Public: %.2f" % (public_adm_count/total_public_count*100) + "%")
```

```python
#Count of 80%+ acceptance of private school applicants (into the UC System)
private_adm_count = all_complete_new[(all_complete_new["Count"] != "App")
                                      & (all_complete_new["Percent of Applicant Pool"] >= 80)
                                      & (all_complete_new["Institution_type"] == "Private")]["Calculation1"].count()

total_private_count = all_complete_new[(all_complete_new["Count"] == "App") &
                  (all_complete_new["Institution_type"] == "Private")]["Calculation1"].count()

private_adm_count/total_private_count*100
print("Private: %.2f" % (private_adm_count/total_private_count*100) + "%")
```
