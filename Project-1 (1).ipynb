```python
#Dependencies
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats

#Files to Load for THE YEAR 2020
public_school_path = "./uc_data1.csv"
private_school_path = "./private_school_data.csv"

#Read the data file
public_school = pd.read_csv(public_school_path)
public_school['Institution_type'] = 'Public'
public_school['Year'] = 2020
private_school = pd.read_csv(private_school_path)
private_school['Institution_type'] = 'Private'
private_school['Year'] = 2020
# Display the data table for preview
uc_data_complete_2020 = pd.concat([public_school, private_school])
uc_data_complete_2020
```

```python
##Files to Load for THE YEAR 2019
public_snine_path = "./public_s_nine.csv"
private_snine_path = "./private_s_nine.csv"

#Read the data file
nine_public = pd.read_csv(public_snine_path)
nine_public['Institution_type'] = 'Public'
nine_public['Year'] = 2019

nine_private = pd.read_csv(private_snine_path)
nine_private['Institution_type'] = 'Public'
nine_private['Year'] = 2019

# Merge & Display the data table for preview
nine_data_complete = pd.concat([nine_public, nine_private])
nine_data_complete
```

```python
#Files to Load for THE YEAR 2018
public_seight_path = "./public_s_eight.csv"
private_seight_path = "./private_s_eight.csv"

#Read the data file
eight_public = pd.read_csv(public_seight_path)
eight_public['Institution_type'] = 'Public'
eight_public['Year'] = 2018

eight_private = pd.read_csv(private_seight_path)
eight_private['Institution_type'] = 'Public'
eight_private['Year'] = 2018

# Merge & Display the data table for preview
eight_data_complete = pd.concat([eight_public, eight_private])
eight_data_complete
```

```python
#Files to Load for THE YEAR 2017
public_sseven_path = "./public_s_seven.csv"
private_sseven_path = "./private_s_seven.csv"

#Read the data file
seven_public = pd.read_csv(public_sseven_path)
seven_public['Institution_type'] = 'Public'
seven_public['Year'] = 2017

seven_private = pd.read_csv(private_sseven_path)
seven_private['Institution_type'] = 'Public'
seven_private['Year'] = 2017

# Merge & Display the data table for preview
seven_data_complete = pd.concat([seven_public, seven_private])
seven_data_complete
```

```python
#Files to Load for THE YEAR 2016
public_ssix_path = "./public_s_six.csv"
private_ssix_path = "./private_s_six.csv"

#Read the data file
six_public = pd.read_csv(public_ssix_path)
six_public['Institution_type'] = 'Public'
six_public['Year'] = 2016

six_private = pd.read_csv(private_ssix_path)
six_private['Institution_type'] = 'Public'
six_private['Year'] = 2016

# Merge & Display the data table for preview
six_data_complete = pd.concat([six_public, six_private])
six_data_complete
 

```

```python
#Merging the data of all 5 years of public and private school data using concat 

All_complete_data = pd.concat([uc_data_complete_2020, nine_data_complete, eight_data_complete, seven_data_complete, six_data_complete])
All_complete_data
```

```python
#Finding the total number of admissions, applications, and enrollment of 2016-2020 (as a whole)

All_complete_data["Count"].value_counts()
```

```python
#Finding the number of Adm, App, Enr for EACH year (make it into series, like a list)
All_complete_data.groupby("Year")["Count"].value_counts()
```

```python
#Convert into a dataframe (turn a series into a table, like a matrix to see the rows and columns)
df = pd.DataFrame (All_complete_data.groupby("Year")["Count"].value_counts())
df
df.rename_axis(index={"Count":"Type"}, inplace = True)
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
##### Side note: we should make the bars a bit more cleaner maybe in percentages?
#pivot data to be able to generate each column of Adm, App, Enr in a bar graph
df.pivot("Year","Type", "Count").plot(kind="bar")
```

```python
#### We need to fix the settings , bascially the overall data visualization for it to be more readable.
# Generate a scatter plot of year versus Adm,App,Enr
year=df["Year"]
y = df["Count"]
plt.scatter(year,y)
plt.xticks(year, rotation=90)
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```
